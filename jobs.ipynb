```python
def extended_infer_domain(text):
    text = str(text).lower()
    if any(word in text for word in ['data', 'analytics', 'machine learning']):
        return 'Data/Analytics'
    elif any(word in text for word in ['it support', 'systems', 'network']):
        return 'IT Support'
    elif any(word in text for word in ['developer', 'engineer', 'software']):
        return 'Software Development'
    elif any(word in text for word in ['project manager', 'scrum', 'agile']):
        return 'Project Management'
    elif any(word in text for word in ['consultant', 'business', 'strategy']):
        return 'Consulting'
    elif any(word in text for word in ['research linguistics', 'syntax', 'phonology']):
        return 'Linguistics'
    elif any(word in text for word in ['nlp', 'natural language', 'language models','llms' ]):
        return 'Computational Linguistics'
    elif any(word in text for word in ['election', 'policy', 'governance', 'public affairs']):
        return 'Political Science'
    elif any(word in text for word in ['marketing', 'branding', 'seo']):
        return 'Marketing'
    elif any(word in text for word in ['journalism', 'reporter', 'editor']):
        return 'Media & Journalism'
    elif any(word in text for word in ['teacher','professor', 'students']):
        return 'Education'
    elif any(word in text for word in ['finance', 'accounting', 'budget']):
        return 'Finance'
    elif any(word in text for word in ['hr', 'recruitment', 'people operations']):
        return 'Human Resources'
    elif any(word in text for word in ['legal', 'attorney', 'compliance']):
        return 'Legal'
    elif any(word in text for word in ['design', 'ux', 'ui', 'illustration']):
        return 'Design & UX'
    elif any(word in text for word in ['healthcare', 'medical', 'clinic']):
        return 'Healthcare'
    elif any(word in text for word in ['logistics', 'supply chain', 'operations']):
        return 'Logistics & Operations'
    elif any(word in text for word in ['restaurant', 'waiter', 'menu', 'cook']):
        return 'Gastronomy & Service'
    elif any(word in text for word in ['hotel', 'travel', 'touristic', 'tourism']):
        return 'Tourism'
    else:
        return 'Other'

data['domain'] = data['job_descript'].apply(extended_infer_domain)

data
```
